```python
# Find customer order

customer_order = ['Mango','Dates','Apple','Pine Apple','Apple','Dates','Mango','Dates','Apple']

order_count = {}

for item in customer_order:
    if item in order_count:
        order_count[item] += 1
    else:
        order_count[item] = 1

print(order_count)
```
